```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Data Science/Time Series/daily-total-female-births.csv'

df = pd.read_csv(path)
```

```python
df.head()
```

```python
feature = df.copy()
feature.head()
```

```python
feature['Date'] =  pd.to_datetime(feature['Date'])
```

```python
feature['year'] = feature['Date'].dt.year
feature['month'] = feature['Date'].dt.month
feature['day'] = feature['Date'].dt.day

feature.head()
```

## Lag Feature

```python
feature['lag1'] = feature['Births'].shift(1) ## here we are creating a lag of Birth column.
feature['lag2'] = feature['Births'].shift(365)

feature.head()  ## The first column in lag1 is NA as nothing is available. Same for lag2 for 365.
```

## Windows Feature

```python
feature['Roll_mean'] = feature['Births'].rolling(window = 2).mean()
feature['Roll_max'] = feature['Births'].rolling(window = 3).max()

feature.head()
```

## Expanding

```python
feature['Expand_max'] = feature['Births'].expanding().max()

feature.head()
```
